```python
import pandas as pd
#importing dataset
df = pd.read_csv('train.csv')
```

```python
df.head()
df.describe()
df.info()           
df.value_counts()
```

```python
#Visualization
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df)  
plt.show()
```

```python
# Heatmap only for numeric columns
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='Blues')
plt.show()
```

```python
#OBSERVATIONS
#Survival by gender
sns.countplot(data=df, x='Sex', hue='Survived', palette='Paired', edgecolor='black')
plt.title('Survival by Gender')
plt.show()
#Survival by Passenger Class
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Paired', edgecolor='black')
plt.title('Survival by Passenger Class')
plt.show()
#Grouped Survival Rates
#Average survival rate by class
print(df.groupby('Pclass')['Survived'].mean())

#Average survival rate by gender
print(df.groupby('Sex')['Survived'].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Age
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='Age', bins=30, kde=True, color='steelblue', edgecolor='black')
plt.title('Age Distribution')
plt.show()

# Boxplot for Fare by Survival
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='Survived', y='Fare', hue='Survived', palette={0: '#add8e6', 1: '#00008b'}, dodge=False, legend=False)
plt.title('Fare Distribution by Survival Status')
plt.show()

# Scatterplot for Age vs Fare colored by Survival
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette={0: '#87ceeb', 1: '#4682b4'})
plt.title('Age vs Fare Scatterplot by Survival')
plt.show()
```
